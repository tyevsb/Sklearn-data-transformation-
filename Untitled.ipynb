```python
#importing all the necessary data
import pandas as pd
car_sales = pd.read_csv("Data/car-sales-extended.csv")
car_sales
```

```python
#spliting the data
x = car_sales.drop("Price", axis = 1)
y = car_sales["Price"]
```

```python
#spliting data into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
#Build machine learning model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
#turning non-numerical datas into number
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
cat_col = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",one_hot,cat_col)],remainder="passthrough")

transformed_x = transformer.fit_transform(x)
transformed_x
```

```python
pd.DataFrame(transformed_x)
```

```python
#spliting data into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(transformed_x,y, test_size = 0.2)
```

```python
#Build machine learning model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test) #the result indicates it hard 
                            #predict the price based on given data
```
